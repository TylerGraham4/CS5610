### We can test for association between categorical and continuous variables using a Kruskal-Wallis test using the Scipy kruskal() function. In this example, we want to know if the distribution of prices for condos is different from the distribution for other property types:

```python
# Create masks to select prices for condos and other property types
condo_mask = real_estate["type"] == "Condo"
other_types_mask = real_estate["type"] != "Condo"

# Select prices for condos and other property types
condo_prices = real_estate["price"][condo_mask]
other_types_prices = real_estate["price"][other_types_mask]
```

```python
statistic, p_value = kruskal(condo_prices, other_types_prices)
print(statistic, p_value)
```

```python
alpha = 0.05  # Significance level
if p_value < alpha:
    print("There is a statistically significant difference in price distributions.")
```

## Part 3

### For each continuous variable, create a box plot of the continuous variable versus property type. Make sure to put the property type on the horizontal axis and the continuous variable on the vertical axis.

```python
property_types = ['Condo', 'Residential', 'Multi-Family']

# Create box plots for each continuous variable versus property type
for variable in continuous_variables:
    plt.figure(figsize=(10, 6))  # Set the figure size
    sns.boxplot(x='type', y=variable, data=real_estate, order=property_types)  # Create the box plot
    plt.title(f'Box Plot of {variable} vs. Property Type')  # Set the plot title
    plt.xlabel('Property Type')  # Label for the x-axis
    plt.ylabel(variable)  # Label for the y-axis
    plt.grid(True)  # Add grid lines
    plt.show()  # Display the plot
```

### Run Kruskal-Wallis test for each continuous variable versus the property type.

```python
variables = []
group1_values = []
group2_values = []
statistics = []
p_values = []
significances = []

# Run Kruskal-Wallis test for each continuous variable versus property type
for variable in continuous_variables:
    for pt1 in property_types:
        for pt2 in property_types:
            if pt1 != pt2:
                group1 = real_estate[real_estate['type'] == pt1][variable]
                group2 = real_estate[real_estate['type'] == pt2][variable]
                statistic, p_value = kruskal(group1, group2)
                significant = 'Yes' if p_value < 0.05 else 'No'
                
                # Append the results to the respective lists
                variables.append(variable)
                group1_values.append(pt1)
                group2_values.append(pt2)
                statistics.append(statistic)
                p_values.append(p_value)
                significances.append(significant)

# Create the DataFrame using the lists
kruskal_results = pd.DataFrame({
    'Variable': variables,
    'Group 1': group1_values,
    'Group 2': group2_values,
    'Statistic': statistics,
    'P-value': p_values,
    'Significant': significances
})

# Display the Kruskal-Wallis test results as a table
print(kruskal_results)
```

### For each categorical variable, create a heat map of the counts of each categorical variable value for each property type.

```python
for variable in categorical_variables:
    # Create a defaultdict to count the values for each property type
    values_type = defaultdict(lambda: defaultdict(int))
    for idx, row in real_estate.iterrows():
        values_type[row['type']][row[variable]] += 1

    # Create a DataFrame from the defaultdict
    heatmap_df = pd.DataFrame.from_dict(values_type)

    # Set the figure size and dpi
    fig = plt.gcf()
    fig.set_dpi(150)

    # Create the heatmap using Seaborn
    sns.heatmap(heatmap_df, annot=True, fmt=".0f", cmap='coolwarm')
    plt.xlabel("Property Type", fontsize=16)
    plt.ylabel(variable, fontsize=16)
    plt.title(f"Heatmap of {variable} Counts by Property Type", fontsize=16)

    plt.show()
```

### Chi Test

```python
for variable in categorical_variables:
    # Create a contingency table of counts between the variable and property type
    contingency_table = pd.crosstab(real_estate[variable], real_estate['type'])

    # Run the Chi-squared test of independence
    chi2, p, _, _ = chi2_contingency(contingency_table)

    # Print the results
    print(f"Chi-squared Test of Independence between {variable} and Property Type:")
    print(f"Chi-squared statistic: {chi2}")
    print(f"P-value: {p}")
    if p < 0.05:
        print("There is a statistically significant association between the variables.")
    else:
        print("There is no statistically significant association between the variables.")
    print("\n")
```

```python
# Divide the square footage by 1000
real_estate['scaled_square_foot'] = real_estate['sq__ft'] / 1000

# Create the design matrix using patsy with no intercept
formula = 'price ~ scaled_square_foot'
y, X = patsy.dmatrices(formula, data=real_estate, return_type='dataframe')

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
```

```python
predicted_prices = model.predict(X)

# Calculate the residuals
residuals = real_estate['price'] - predicted_prices

# Calculate the squared errors
squared_errors = residuals ** 2

# Calculate the mean squared error (MSE)
mse = np.mean(squared_errors)

# Calculate the RMSE
rmse = np.sqrt(mse)

print(f"Root Mean Square Error (RMSE): {rmse}")
```
