### Import Libraries

```python
import tkinter as tk
import pandas as pd
import joblib
```

### Import model and get everything ready for user input

```python
df = pd.read_csv('final_dataset.csv')
df.head(100)
```

```python
disease_class_unique = df['Disease Class'].unique().tolist()
encoded_class_unique = df['encoded_disease_class'].unique().tolist()

print(disease_class_unique)
print(encoded_class_unique)
```

```python
df_pred = df.drop('Disease',axis=1)
df_pred = df_pred.drop('encoded_disease',axis=1)
df_pred = df_pred.drop('Disease Class',axis=1)
df_pred = df_pred.drop('encoded_disease_class',axis=1)
```

```python
single_row = df_pred.iloc[[1]]
single_row.loc[1] = 0
single_row.head()
```

```python
model = joblib.load('LogRegModel.pkl')
```

### Run the GUI

```python
def chatbot_response(user_input):
    if(user_input == "Yes"):
        prediction_num = model.predict(single_row)
        index = encoded_class_unique.index(prediction_num)
        prediction = disease_class_unique[index]
        single_row.loc[1] = 0
        
        return (f'From our analysis, it seem like you may have: {prediction}\n\nChatbot: Welcome to ChatMD! Please enter your symptoms (one at a time) for a diagnosis.\n')
    
    else:
        if user_input in single_row.columns:
            single_row.loc[1,user_input] = 1
            return "Symptom Recorded! Are you done?"
        else:
            return "Couldn't find symptom, please try again!"
```

```python
root = tk.Tk()
root.title("ChatMD")

# Create the message box
text_area = tk.Text(root, bg="white", width=50, height=20)
text_area.pack()

# Create a vertical scroll bar and link with message box
vertical_scrollbar = tk.Scrollbar(root, command=text_area.yview)
vertical_scrollbar.pack(anchor=tk.E,fill=tk.Y)
text_area.config(yscrollcommand=vertical_scrollbar.set)

# Create the user's input field
input_field = tk.Entry(root, width=50)
input_field.pack()

# Create the send button
send_button = tk.Button(root, text="Send", command=lambda: send_message())
send_button.pack()

# Enter welcome message
text_area.insert(tk.END,'Chatbot: Welcome to ChatMD! Please enter your symptoms (one at a time) for a diagnosis.')

def send_message():
  # Get the user's input
  user_input = input_field.get()

  # Clear the input field
  input_field.delete(0, tk.END)

  # Generate a response from the chatbot
  response = chatbot_response(user_input)

  # Display the response in the chatbot's text area
  text_area.insert(tk.END,f"\nUser: {user_input}\n")
  text_area.insert(tk.END,f"\nChatbot: {response}\n")

root.mainloop()
```
