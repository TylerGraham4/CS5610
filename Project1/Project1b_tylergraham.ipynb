### Question 3: Plot the zip codes as a bar (count) plot.  Even if the number of unique zip codes is large, why do you think it is better to use a categorical variable for zip code instead of a numerical variable?

Zip codes are better to be categorical data as opposed to numerical since the numerical value of the zip code is meaningless, it is meant to be another label for a geographical location. Similarly to "state" which also is a label for a geographical location and is a good categorical data. Please see below a graph of the zip codes. 

```python
plt.figure(figsize=(13, 9))
real_estate.zip.value_counts().plot(kind='bar')
plt.title('Distribution of Zip Codes')
plt.xlabel('Zip Code')
plt.ylabel('Count')
plt.show()
```

### Question 4: Convert the variables you think would be appropriate categorical variables to the "category" type. Run the info() method again to verify that these variables have the Dtype "category" next to them.

Convert the columns that are identified as categorical to the category data type

```python
real_estate['city'] = real_estate.city.astype('category')
real_estate['zip'] = real_estate.zip.astype('category')
real_estate['state'] = real_estate.state.astype('category')
real_estate['beds'] = real_estate.beds.astype('category')
real_estate['baths'] = real_estate.baths.astype('category')
real_estate['type'] = real_estate.type.astype('category')
```

Verify with the info() call on the dataset

### Question 2: Come up with criteria to identify records with odd values.  Remove them from the data set.  (Hint: using Pandas' selecting and masking syntax)

The pricing data below $5000 and the 0 square foot data will be removed since they are odd data and inaccurate as described in the above comment sections. 

```python
mask = real_estate['price'] > 5000
new_real_estate = real_estate[mask]
```

```python
mask2 = new_real_estate['sq__ft'] != 0
new_real_estate = new_real_estate[mask2]
new_real_estate.describe()
```
