### Question 1: Load the three data sets.  Create a feature matrix for each from the sepal length and width columns.  Use the LabelEncoder from scikit-learn to convert the string labels to integers.  

```python
# Load the datasets
setosa_training = pd.read_csv("setosa_training.csv")
setosa_target = pd.read_csv("setosa_target.csv")
setosa_testing = pd.read_csv("setosa_testing.csv")

# Create feature matrices
X_training = setosa_training[["sepal_length (cm)", "sepal_width (cm)"]].values
X_target = setosa_target[["sepal_length (cm)", "sepal_width (cm)"]].values
X_testing = setosa_testing[["sepal_length (cm)", "sepal_width (cm)"]].values
```

```python
# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode the class labels
y_training= label_encoder.fit_transform(setosa_training["label"])
y_target = label_encoder.fit_transform(setosa_target["label"])
y_testing = label_encoder.fit_transform(setosa_testing["label"])
```

### Question 2: Train a logistic regression model on the training set using the LogisticRegression class from scikit-learn.  Generate predictions for the training, testing, and target sets and evaluate the predictions by computing the accuracy scores.

```python
# Train Logistic Regression models
model_training = LogisticRegression()
model_training.fit(X_training, y_training)

# Generate predictions for training sets
y_training_train_pred = model_training.predict(X_training)
y_target_train_pred = model_training.predict(X_target)
y_testing_train_pred = model_training.predict(X_testing)

# Calculate accuracy scores for training sets
accuracy_training = accuracy_score(y_training, y_training_train_pred)
accuracy_target = accuracy_score(y_target, y_target_train_pred)
accuracy_testing = accuracy_score(y_testing, y_testing_train_pred)

print(f"Accuracy Training: {accuracy_training}")
print(f"Accuracy Target: {accuracy_target}")
print(f"Accuracy Testing: {accuracy_testing}")
```

### Question 3: You can extract the learned values for the decision boundary from the trained model using the coef_ and intercept_ attributes.  The two values in coef_ correspond to A and B and intercept_ refers to C in the equation form Ax + By + C.  Plot the learned decision boundary on scatter plots for the three data sets with the points colored by class.

```python
# Extract coefficients and intercept for decision boundary
coef_training = model_training.coef_[0]
intercept_training = model_training.intercept_[0]
```

```python
plt.scatter(X_training[:, 0], X_training[:, 1], c=y_training, cmap='viridis', label='label')
x_values = np.linspace(4, 8, 100)
y_values = (-coef_training[0] / coef_training[1]) * x_values - intercept_training / coef_training[1]
plt.plot(x_values, y_values, '-r', label='Decision Boundary')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.title('Setosa Training Data with Decision Boundary')
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(X_target[:, 0], X_target[:, 1], c=y_target, cmap='viridis', label='label')
x_values = np.linspace(4, 8, 100)
y_values = (-coef_training[0] / coef_training[1]) * x_values - intercept_training / coef_training[1]
plt.plot(x_values, y_values, '-r', label='Decision Boundary')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.title('Setosa Target Data with Decision Boundary')
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(X_testing[:, 0], X_testing[:, 1], c=y_testing, cmap='viridis', label='label')
x_values = np.linspace(4, 8, 100)
y_values = (-coef_training[0] / coef_training[1]) * x_values - intercept_training / coef_training[1]
plt.plot(x_values, y_values, '-r', label='Decision Boundary')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.title('Setosa Testing Data with Decision Boundary')
plt.show()
```
