### Question 4: Plot in the specified color and label the decision boundaries defined by the following equations:

Decision Boundary 1 (black): y = 2x – 8
Decision Boundary 2 (red): y = 0.75x – 0.9
Decision Boundary 3 (purple): y = -2x + 12

```python
# Sample data
x = setosa_df['sepal_length (cm)']
y = setosa_df['sepal_width (cm)']

# Create a scatter plot
sns.scatterplot(x=x, y=y, data=setosa_df, alpha=0.5, hue='label', palette={"setosa": "orange", "not setosa": "blue"})
plt.title(f'Scatter Plot of Sepal Length vs. Sepal Width')

# Add labels
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

# Plot the line y = 2x - 8 in red with a dashed line style
plt.plot(x, 2 * x - 8, color='black', linestyle='--', label='y = 2x - 8')
plt.plot(x, .75 * x - 9, color='red', linestyle='--', label='y = .75x - 8')
plt.plot(x, -2 * x + 12, color='purple', linestyle='--', label='y = -2x - 8')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

### Question 5: Rewrite each decision boundary equation in a modified standard form: Ax + By + c = 0

```python
boundary1 = [2.0, -1.0, -8.0]
boundary2 = [0.75, -1.0, -0.9]
boundary3 = [-2.0, -1.0, 12.0]
```

### Question 6: Import the linear_decision_boundary_classifier() function from the provided decision_boundaries.py file.  Call the function for each decision boundary like so:
### Question 7: Use the scikit-learn sklearn.metrics.accuracy_score() function to calculate and print the accuracy of the predictions using each decision boundary using the true and predicted labels.

```python
true_labels = setosa_df["label"]

features = setosa_df[["sepal_length (cm)", "sepal_width (cm)"]].values
class_names = ["not setosa", "setosa"]
pred_labels = linear_decision_boundary_classifier(boundary1, features, class_names=class_names)
print(pred_labels)
accuracy = accuracy_score(true_labels, pred_labels)
print(f"Accuracy for Decision Boundary 1: {accuracy:.2f}")

# Boundary 2
features = setosa_df[["sepal_length (cm)", "sepal_width (cm)"]].values
class_names = ["not setosa", "setosa"]
pred_labels = linear_decision_boundary_classifier(boundary2, features, class_names=class_names)
print(pred_labels)
accuracy = accuracy_score(true_labels, pred_labels)
print(f"Accuracy for Decision Boundary 2: {accuracy:.2f}")

# Boundary 3
features = setosa_df[["sepal_length (cm)", "sepal_width (cm)"]].values
class_names = ["not setosa", "setosa"]
pred_labels = linear_decision_boundary_classifier(boundary3, features, class_names=class_names)
print(pred_labels)
accuracy = accuracy_score(true_labels, pred_labels)
print(f"Accuracy for Decision Boundary 3: {accuracy:.2f}")# Boundary 1
```
